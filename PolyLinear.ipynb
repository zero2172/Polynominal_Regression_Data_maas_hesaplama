```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
data=pd.read_csv("polynomial.csv",sep=";")
data
```

```python
plt.plot(data.deneyim,data.maas,"r.-",label="maaş")
plt.legend()
plt.xlabel("deneyim")
plt.ylabel("maaş")
```

```python
reg=LinearRegression()
reg.fit(data[["deneyim"]], data["maas"])
plt.xlabel("deneyim")
plt.ylabel("maaş")

plt.plot(data["deneyim"],data["maas"],color="red")

xekseni=data["deneyim"]
yekseni=reg.predict(data[["deneyim"]])
plt.plot(xekseni, yekseni,"b.-",label="Linear regression")
plt.legend();
```

```python
polynominal_regresson=PolynomialFeatures(degree=4)
x_polynomial=polynominal_regresson.fit_transform(data[["deneyim"]])
```

```python
reg=LinearRegression()
reg.fit(x_polynomial,data["maas"])
```

```python
y_head=reg.predict(x_polynomial)
plt.plot(data["deneyim"],y_head,color="red",label="polynomial regression")
plt.legend()
plt.scatter(data["deneyim"], data["maas"]);
```

```python
x_polyreg=polynominal_regresson.fit_transform([[4.5]])
reg.predict(x_polyreg)
```
